```python
import matplotlib.pyplot as plt
```

```python
#Delta Function
def delta(d):
    plt.figure(figsize=(10, 5))
    for n in range(-10, 10):
        if d(n) == 0:
            delta = 1
        else:
            delta = 0
        plt.stem(n, delta)

    plt.title('Delta Function')
    plt.xlabel('n')
    plt.ylabel('delta[n]')
    plt.xticks(range(-10, 10))

```

```python
# Unit Step Function
def unit(d):
    plt.figure(figsize=(10, 5))
    for n in range(-10, 10):
        if d(n) < 0:
            u = 0
        else:
            u = 1
        plt.stem(n, u)

    plt.title('Unit Step Function')
    plt.xlabel('n')
    plt.ylabel('u[n]')
    plt.xticks(range(-10, 10))
```

```python
# Ramp Function
def ramp(d):
    plt.figure(figsize=(10, 5))
    for n in range(-10, 10):
        if d(n) < 0:
            p = 0
        else:
            p = n
        plt.stem(n, p)

    plt.title('Ramp Function')
    plt.xlabel('n')
    plt.ylabel('p[n]')
    plt.xticks(range(-10, 10))
```

```python
# (2^n)*u[n]
def exp(d):
    plt.figure(figsize=(10, 5))
    for n in range(-10, 10):
        if d(n) < 0:
            exp = 0
        else:
            exp = 2**n
        plt.stem(n, exp)

    plt.title('Exponential Function')
    plt.xlabel('n')
    plt.ylabel('2^n u[n]')
    plt.xticks(range(-10, 10))
```

```python
#Delta Function Graph
delta(lambda n: n)
```

```python
#Unit Step Function Graph
unit(lambda n: n)
```

```python
#Ramp Function Graph
ramp(lambda n: n)
```

```python
# (2^n)*u[n] Function Graph
exp(lambda n: n)
```

```python
# (2^n)*u[n+1] Function Graph
exp(lambda n: n+1)
```

```python
# (2^n)*u[n-1] Function Graph
exp(lambda n: n-1)
```

```python
# (2^n)*u[-n+1] Function Graph
exp(lambda n: -n+1)
```

```python
# (2^n)*u[-n-1] Function Graph
exp(lambda n: -n-1)
```
